```python
#@title **Step 1 : Prepare Files**

# #@title **User Login**

import os

os.system("pip install bcrypt > /dev/null 2>&1")
from IPython.display import clear_output

# User input for username and password using Colab's param feature
username = 'user1'  #param {type:"string"}
password = '123'  #param {type:"string"}

# Function to verify user credentials
import sqlite3
import hashlib

# Connect to SQLite database
file_id = "1L6EIBl8WEzrPJw3C3AmlUTCACqCgYcKY"
destination = "/content/users.db"
os.system(f"gdown --id {file_id} -O {destination} > /dev/null 2>&1")
conn = sqlite3.connect('/content/users.db')
cursor = conn.cursor()

# Function to verify user credentials
def verify_user(username, password):
    """Verify the username and password."""
    cursor.execute('SELECT * FROM users WHERE username = ?', (username,))
    user = cursor.fetchone()

    if user:
        stored_hash = user[2]  # Get the stored password hash
        entered_hash = hashlib.sha256(password.encode()).hexdigest()
        if entered_hash == stored_hash:
            return True
        else:
            return False
    else:
        return False

# Authenticate user
if verify_user(username, password):
    # print(f"✅ Access granted for {username}!")

    # --- NEW RVC DOWNLOAD VIA HUGGING FACE ---
    import zipfile

    hf_rvc_zip = "https://huggingface.co/datasets/BahaaMahmoud88/testrvc/resolve/main/RVC.zip"
    zip_path = "/content/RVC.zip"
    os.system(f"wget -O {zip_path} {hf_rvc_zip}")

    # Validate download
    if not os.path.exists(zip_path) or os.path.getsize(zip_path) < 5_000_000:
        raise FileNotFoundError("❌ RVC.zip is missing or corrupted!")

    # Extract zip
    extract_path = "/content/RVC"
    os.makedirs(extract_path, exist_ok=True)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)

    print(f"✅ RVC extracted to {extract_path}")

    # Optional: remove zip after extraction
    if os.path.exists(zip_path):
        os.remove(zip_path)

    # Remove users.db after successful login
    db_path = "/content/users.db"
    if os.path.exists(db_path):
        os.remove(db_path)

    # Save credentials to a JSON file
    import json
    credentials = {'username': username, 'password': password}
    os.makedirs('/content/RVC/infer/modules/train', exist_ok=True)
    with open('/content/RVC/infer/modules/train/credentials.json', 'w') as f:
        json.dump(credentials, f)
else:
    print("❌ Access denied! Invalid username or password.")
```

```python
#@title **Step 3 :** Download New Song

import os
from pydub import AudioSegment
from google.colab import files

# Define the path for the new song
new_song_path = '/content/sample_data/new_song.wav'

# Delete the file if it exists
if os.path.exists(new_song_path):
    os.remove(new_song_path)
    print(f"{new_song_path} deleted.")
else:
    print(f"{new_song_path} does not exist.")

# Define file paths
vocals_path = "/content/RVC/audios/output_audio.wav"
music_path = "/content/sample_data/music.wav"
output_path = "/content/sample_data/new_song.wav"

# Load the vocals and music files
vocals = AudioSegment.from_file(vocals_path)
music = AudioSegment.from_file(music_path)

# Mix the two audio tracks together
mixed = vocals.overlay(music)

# Export the mixed song
mixed.export(output_path, format="wav")
print(f"Mixed song saved as {output_path}")

# Download the mixed song
files.download(output_path)



```
